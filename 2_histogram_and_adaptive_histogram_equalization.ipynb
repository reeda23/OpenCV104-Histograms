```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#func to show image
def plt_showimage(title, image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

```python
#read image from the disk
image = cv2.imread("/content/drive/MyDrive/OpenCV104 - Histograms/beach.png")
#convert it into gray scale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Histogram equalization
equlaized = cv2.equalizeHist(gray)
```

```python
plt_showimage("Input Image", gray)
plt_showimage("Equalized Image", equlaized)
```

**Adaptive Histogram Equalization**

```python
args = {
    "image":"/content/drive/MyDrive/OpenCV104 - Histograms/beach.png",
    "clip" : 2.0,
    "tile" : 8
}
```

```python
#load the input image and convert it into gray scale

image = cv2.imread(args["image"])
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# apply CLAHE (Contrast Limited Adaptive Histogram Equalization)
clahe = cv2.createCLAHE(clipLimit = args["clip"], tileGridSize = (args["tile"], args["tile"]))
equalized = clahe.apply(gray)
```

```python
# show the original grayscale image and CLAHE output image
plt_showimage("Input", gray)
plt_showimage("CLAHE", equalized)
```
